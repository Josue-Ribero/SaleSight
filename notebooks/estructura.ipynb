# 📦 Estructura del Proyecto - SaleSight

Este proyecto sigue la estructura estándar de **Cookiecutter Data
Science**, organizada para proyectos de análisis y modelos de Machine
Learning.

------------------------------------------------------------------------

## 📁 entorno/

Carpeta del entorno virtual de Python.\
Contiene las dependencias instaladas del proyecto.\
⚠️ No se sube al repositorio.

------------------------------------------------------------------------

## 📁 data/

Almacena todos los datos del proyecto.

-   **external/** → Datos provenientes de fuentes externas.
-   **interim/** → Datos intermedios después de limpieza parcial.
-   **processed/** → Datos finales listos para modelado.
-   **raw/** → Datos originales sin modificar.

------------------------------------------------------------------------

## 📁 docs/

Documentación del proyecto.

-   **docs/** → Archivos adicionales de documentación.
-   **mkdocs.yml** → Configuración de documentación con MkDocs.
-   **README.md** → Documentación general del proyecto.

------------------------------------------------------------------------

## 📁 models/

Contiene modelos entrenados o serializados.

-   **.gitkeep** → Archivo vacío para que Git mantenga la carpeta.

------------------------------------------------------------------------

## 📁 notebooks/

Jupyter Notebooks para análisis exploratorio y experimentación.

-   **estructura.ipynb** → Notebook para explicar la estructura de carpetas
-   **.gitkeep** → Mantiene la carpeta en el repositorio.

------------------------------------------------------------------------

## 📁 references/

Material de referencia como diccionarios de datos, papers o
documentación externa.

------------------------------------------------------------------------

## 📁 reports/

Resultados generados.

-   **figures/** → Gráficas generadas por el modelo o análisis.
-   **.gitkeep** → Mantiene la carpeta en Git.

------------------------------------------------------------------------

## 📁 salesight/

Paquete principal de Python del proyecto.

### 📁 modeling/

Código relacionado con entrenamiento y evaluación de modelos.

-   ****init**.py** → Indica que la carpeta es un módulo Python.
-   **config.py** → Configuración del modelo (parámetros, rutas,
    constantes).
-   **dataset.py** → Funciones para carga y preparación de datos.
-   **features.py** → Ingeniería de características.
-   **plots.py** → Funciones para visualización de datos.

------------------------------------------------------------------------

## 📁 tests/

Pruebas unitarias del proyecto.

-   **test_data.py** → Pruebas relacionadas con el procesamiento de
    datos.

------------------------------------------------------------------------

## Archivos raíz del proyecto

-   **.env** → Variables de entorno (configuración sensible).
-   **.gitignore** → Define qué archivos no se suben a Git.
-   **Makefile** → Automatización de tareas (limpieza, entrenamiento,
    etc.).
-   **pyproject.toml** → Configuración del proyecto y dependencias.
-   **README.md** → Descripción general del proyecto.
-   **requirements.txt** → Lista de dependencias del proyecto.

------------------------------------------------------------------------

# 🎯 Flujo típico del proyecto

1.  Se colocan datos crudos en `data/raw/`
2.  Se procesan y limpian hacia `data/interim/`
3.  Se generan datos listos para modelado en `data/processed/`
4.  Se entrenan modelos desde `salesight/modeling/`
5.  Se guardan resultados en `models/`
6.  Se generan reportes y figuras en `reports/`

------------------------------------------------------------------------

Estructura diseñada para mantener organización, reproducibilidad y
buenas prácticas en proyectos de Ciencia de Datos.
